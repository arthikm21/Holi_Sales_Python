```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Holi Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status'], axis = 1, inplace=True)
```

```python
df.info()
```

```python
df.drop(['unnamed1'], axis = 1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
##EXPLORATORY DATA ANALYSIS
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_age)
```

```python
#AGE
```

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
 ax.bar_label(bars)
```

```python
# total number of orders from 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(26,8)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
# total number of orders from 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)

sns.set({'figure.figsize':(27,9)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
            
```

```python
ax = sns.countplot(data=df, x='Marital_Status')

# Adding labels to the bars
sns.set(rc={'figure.figsize':(21,5)})
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
```

```python
#MARITAL STATUS VS AMOUNT
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.set(rc={'figure.figsize':(8,7)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
                                                                                            
```

```python
#OCCUPATION
```

```python
ax = sns.countplot( data = df, x = 'Occupation')
sns.set(rc={'figure.figsize':(29,9)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')
sns.set(rc={'figure.figsize':(29,9)})
```

```python
*From above graph, we can figure out that most of the buyers are working in IT, healthcare and Aviation*
```

```python
#PRODUCT CATEGORY
```

```python
ax = sns.countplot(data = df, x = 'Product_Category')
sns.set({'figure.figsize':(29,9)})
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(29,9)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')
```

```python
#From above graph, most of the sold products are from food, clothing and electronics category
```

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
